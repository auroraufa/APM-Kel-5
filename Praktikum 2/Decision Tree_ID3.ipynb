```python
# import libraries
import pandas as pd # untuk menggunakan pandas, import pandas sebagai pd
from sklearn.preprocessing import LabelEncoder # untuk transformasi ke dalam bentuk numerik
from sklearn.model_selection import train_test_split # untuk memprediksi tahapan dan kinerja algortima
from sklearn.metrics import accuracy_score, recall_score, precision_score # untuk menghitung akurasi
from sklearn.tree import DecisionTreeClassifier # untuk membuat model decision tree classification
import six
import sys
sys.modules['sklearn.externals.six']=six
from id3 import Id3Estimator
from id3 import export_graphviz
import numpy as np
```

```python
# membaca data
data = pd.read_csv("healthcare-dataset-stroke-data.csv")
data
```

```python
# mengubah data menjadi numerik
enc = LabelEncoder()
data['gender'] = enc.fit_transform(data['gender'].values)
data['hypertension'] = enc.fit_transform(data['hypertension'].values)
data['heart_disease'] = enc.fit_transform(data['heart_disease'].values)
data['ever_married'] = enc.fit_transform(data['ever_married'].values)
data['work_type'] = enc.fit_transform(data['work_type'].values)
data['Residence_type'] = enc.fit_transform(data['Residence_type'].values)
data['smoking_status'] = enc.fit_transform(data['smoking_status'].values)
data['bmi'] = enc.fit_transform(data['bmi'].values)
data.head()
```

```python
data.info()
```

```python
# menentukan label dari dataset
atr_data = data.drop(columns='stroke') # mengklasifikasi kolom stroke
atr_data.head()
```

```python
# melihat data label
cls_data = data['stroke']
cls_data
```

```python
# xtrain (fitur training), xtest (fitur testing), ytrain(label untuk data training), ytest (kolom untuk data testing)
xtrain, xtest, ytrain, ytest = train_test_split(atr_data, cls_data, test_size=0.2, random_state=500)
tree_data = DecisionTreeClassifier(random_state=500) # untuk memanggil data klasifikasi
tree_data.fit(xtrain, ytrain)
```

```python
# ID3
# membuat classifier dulu dimana kita akan menurunkan function dari sklearn.tree yg bernama DecisionTreeClassifier
# membuat variabel baru bernama Dtree dan mengimport Tree serta mewariskan class DecisionTreeClassifier
# melakukan proses inherit function fit yg berguna untuk melakukan proses training data menggunakan Decision Tree
Dtree = DecisionTreeClassifier()
Dtree.fit(xtrain, ytrain)
```

```python
# setelah melakukan proses training, selanjutnya melakukan prediksi dengan membandingkan dengan data test dan mengolahnya
predik = Dtree.predict(xtest)
print("sensitifitas :", recall_score(ytest, predik))
print("akurasi :", accuracy_score(ytest, predik))
print("precision :", precision_score(ytest, predik))
```

```python
# membuat klasifikasi menggunakan metode algoritma id3 dengan cara 
# deklarasi ID3Estimator
clf = Id3Estimator()
# melakukan pelatihan dengan metode id3
clf.fit(xtrain, ytrain, check_input=True)
# export data ke format DOT graphviz
export_graphviz(clf.tree_, "data.dot", feature_names=atr_data.columns)

# membuat sebuah variabel baru bernama clf dan mengimport ID3 serta mewariskan class bernama Id3Estimator, 
# kemudian melakukan proses inherit function fit yg berguna untuk melakukan proses training data lalu
# mengexport data yg sudah ditraining dalam bentuk data Tree yg berekstensi .DOT
```
